###`Q-5:` In IPL matches dataset some teams name has changed.
You will have to consider them as same.
```
'Delhi Capitals' formerly as 'Delhi Daredevils'
'Punjab Kings' formerly as 'Kings XI Punjab'
'Rising Pune Supergiant' formerly as 'Rising Pune Supergiants'
```
You need to make changes accordingly. Consider current name for each teams.

Be careful Gujrat Titans and Gujrat Lions are different teams.

````python
# code here
ipl=pd.read_csv(r'C:\Users\alok7\Downloads\IPL_Matches_2008_2022.csv')
team_name_changes={
    'Delhi Daredevils':'Delhi Capitals',
   'Punjab Kings':'Kings XI Panjab',
'Rising Pune Supergiant':'Rising Pune Supergiants'
}
ipl['Team1']=ipl['Team1'].replace(team_name_changes)
ipl['Team2']=ipl['Team2'].replace(team_name_changes)
ipl['Winning']=ipl['WinningTeam'].replace(team_name_changes)
ipl['TossWinner']=ipl['TossWinner'].replace(team_name_changes)
ipl
````

###`Q-6` Write a code which can display the bar chart of top 5 teams who have played maximum number of matches in the IPL.
>  Hint: Be careful the data is divided in 2 different cols(Team 1 and Team 2)


````python
# code here
import matplotlib.pyplot as plt
matches_played = ipl['Team1'].append(ipl['Team2']).value_counts()
top_5_teams= matches_played.head(5)
top_5_teams.plot(kind='bar',color='skyblue')
plt.xlabel('Teams')
plt.ylabel('Number of Matches Played')
plt.title('Top 5 Teams With Maximum Matches Played in IPL')
plt.show()
# Check for NaN values in 'Team1' and 'Team2' columns
# nan_team1 = ipl['Team1'].isna().sum()
# nan_team2 = ipl['Team2'].isna().sum()

# print(f"NaN values in 'Team1': {nan_team1}")
# print(f"NaN values in 'Team2': {nan_team2}")

````
